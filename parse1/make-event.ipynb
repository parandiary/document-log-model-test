```python
import time
import random
import psycopg2
from datetime import datetime, timedelta

# PostgreSQL 데이터베이스 연결 설정
DB_CONFIG = {
    'dbname': 'logdb',
    'user': 'admin',
    'password': 'admin',
    'host': 'localhost',
    'port': 5432
}

# 이벤트 목록 (랜덤 선택용)
EVENT_TYPES = ['다운로드', '삭제', '수정', '열람']
IP_ADDRESSES = ['192.168.0.10', '192.168.0.15', '192.168.0.20', '10.0.0.1', '10.0.0.5']
USER_IDS = ['U001', 'U002', 'U003', 'U004', 'U005']
DOCUMENT_IDS = ['D001', 'D002', 'D003', 'D004', 'D005', 'D006', 'D007', 'D008']

# PostgreSQL에 이벤트 삽입 함수
def insert_event():
    try:
        # PostgreSQL 연결
        conn = psycopg2.connect(**DB_CONFIG)
        cursor = conn.cursor()

        # 랜덤 이벤트 생성
        user_id = random.choice(USER_IDS)
        document_id = random.choice(DOCUMENT_IDS)
        activity_type = random.choice(EVENT_TYPES)
        ip_address = random.choice(IP_ADDRESSES)

        # 현재 시간 설정 (ISO 8601 형식)
        timestamp = datetime.utcnow().isoformat()

        # INSERT 쿼리 작성 및 실행
        query = """
            INSERT INTO document_logs (user_id, document_id, activity_type, timestamp, ip_address)
            VALUES (%s, %s, %s, %s, %s)
        """
        cursor.execute(query, (user_id, document_id, activity_type, timestamp, ip_address))

        # 커밋 및 연결 종료
        conn.commit()
        print(f"Event generated: {user_id}, {document_id}, {activity_type}, {timestamp}, {ip_address}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        if conn:
            cursor.close()
            conn.close()

# 이벤트 주기적으로 생성 (10초 간격)
def generate_events():
    while True:
        insert_event()
        time.sleep(10)  # 10초 대기 후 다음 이벤트 생성

# 이벤트 생성 시작
if __name__ == "__main__":
    generate_events()
```
